```python
import pandas as pd # pandas representated as pd (typical in programming)
import numpy as np # numpy representated as np (typical in programming)
import src
import src.data_converter as converter
from src.model import DecisionTree # decision tree calls
# from sklearn.model_selection import train_test_split
#from sklearn.metrics import confusion_matrix # confusion matrix module
#import seaborn as sns # to plot confusion matrix
import matplotlib.pyplot as plt # to plot confusion matrix
import pprint #prints tree representation in a neat format

data = pd.read_csv("data/car_evaluation.csv", names= ["buying price", "maintenance cost", "number of doors", "number of persons", "lug_boot", "safety", "decision"])

# train test split of data
shuffled_data = data.sample(frac = 1, random_state=42)

# Define a size for your train set 
train_size = int(0.9 * len(data))

# Split your dataset 
train_df = shuffled_data[:train_size]
test_df = shuffled_data[train_size:]
train = converter.dataframe_to_dict(train_df)
test = converter.dataframe_to_dict(test_df)

label = {"decision": train["decision"]}
features = train
del features["decision"]
train_data = {"features": features, "label": label}

tree = DecisionTree(7)
d = {'buying price': train_data["features"]['buying price'],
 'maintenance cost': train_data["features"]['maintenance cost'],
 'number of doors': train_data["features"]['number of doors'], 
 'number of persons': train_data["features"]['number of persons'],
 'lug_boot':train_data["features"]['lug_boot'],
 'safety': train_data["features"]['safety']}

l = {'decision':train_data["label"]['decision']}

m = tree.fit(d, l)
# print(tree.representation)

prediction_list = []
true_list = []
accuracy_list = []
correct_list = []
wrong_list = []
for i in range(len(test_df)):
    feature_record = dict(test_df.iloc[i])
    label = feature_record["decision"]
    del feature_record["decision"]
    prediction = tree.predict(feature_record)
    prediction_list.append(prediction)
    true_list.append(label)
    correct_list.append(label)
    if label == prediction:
        accuracy_list.append(1)
    else:
        accuracy_list.append(0)

prediction_list = []
true_list = []
accuracy_list = []
correct_list = []
wrong_list = []
for i in range(len(test_df)):
    feature_record = dict(test_df.iloc[i])
    label = feature_record["decision"]
    del feature_record["decision"]
    prediction = tree.predict(feature_record)
    prediction_list.append(prediction)
    true_list.append(label)
    correct_list.append(label)
    if label == prediction:
        accuracy_list.append(1)
    else:
        accuracy_list.append(0)

print("accuracy", sum(accuracy_list)/ len(accuracy_list))
```
